## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.head()
```

#### График зависимости стоимости недвижимости от жилой квадратуры

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg', height=5);
```

На графике видно, что чем больше жилая квадратура `sqft_living`, тем выше стоимость `price`.

#### График зависимости стоимости недвижимости от количества комнат

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg', height=5);
```

На графике видно, что чем больше количество комнат `bedrooms`, тем выше стоимость `price`.

#### График зависимости стоимости недвижимости от вида

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
```

На графике видно, что наибольшую стоимость имеют дома с оценкой вида 4.

#### График зависимости стоимости недвижимости от вида на набережную

```python
sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('waterfront')
```

На графике видно, что наибольшую стоимость имеют дома с видом на набережную.

#### График зависимости стоимости недвижимости от состояния дома

```python
sns.boxplot(x=df['price'], y=df['condition'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('condition')
```

На графике видно, что чем лучше состояние дома `condition`, тем выше его стоимость `price`.
